**TASK - 5**

Check for Missing values and deal with them

```python
pg.isnull().any()
```

```python
#For numerical column - null values replacement

pg['culmen_length_mm'].fillna(pg['culmen_length_mm'].median(),inplace =True)
pg['culmen_depth_mm'].fillna(pg['culmen_depth_mm'].median(),inplace =True)
pg['flipper_length_mm'].fillna(pg['flipper_length_mm'].median(),inplace =True)
pg['body_mass_g'].fillna(pg['body_mass_g'].median(),inplace =True)
```

```python
#For categorical collumn - null values replacement

pg['sex'].fillna(pg['sex'].mode(),inplace =True)
```

```python
pg.isnull().any()
```

```python
pg['sex'].replace('.', 'MALE', inplace=True)
```

```python
pg.sex.value_counts()
```

**TASK - 7**

Check the correlation of independent variables with the target

```python
target_column = 'species'
independent_columns = ['island', 'culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
correlation_matrix = pg[independent_columns + [target_column]].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation with Species (Target Variable) Heatmap')
plt.show()
```

**TASK - 8**

Check for Categorical columns and perform encoding

```python
#Lable Encoding

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
pg.island = le.fit_transform(pg.island)
pg.sex = le.fit_transform(pg.sex)
```

**TASK - 9**

Split the data into dependent and independent variables

```python
dependent_variable = 'species'
independent_variables = ['island', 'culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'body_mass_g']
dependent = pg[independent_variables]
independent = pg[dependent_variable]
print("Independent Variables:")
print(independent.head())

print("\nDependent Variable:")
print(dependent.head())
```
